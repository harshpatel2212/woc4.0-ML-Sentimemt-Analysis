#**Data Cleaning**

There are 2 ways to do data cleaning
+ Data filling 
+ Drop off the data

#**Imp Tips**
+ Droping of data is not prefered as we are going to drop some valuable data. 
+ If you already have huge amout of data then droping of data is last option preferred.
+ So, data filling is genereally used and important to learn and so we'll use pandas inbuilt datafilling techniques like: bfill, ffill, etc.

We'll now see both the techniques one by one.
1.   Drop Na values


```python
melb_df_drop = melb_df_copy.copy(deep = True)
```

```python
melb_df_drop.dropna(inplace = True)
```

2. Data Filling 

+ backfill/ bfill - fills the empty cell with the next valid value.
+ pad/ ffill - fills the empty cell with the last valid value encountered till now.

+ Other standard technique is filling the categorical data with the modal value and numeric data with mean/ median value. This technique is used to clean particular feature of the data. 

```python
melb_df_copy.fillna(method = 'bfill', inplace = True)
```

For example variable **Car** can be filled using the **modal** value as its a categorical feature(though it has integral enteries of the number of cars). We can fill the column like **Price** with the **mean/ median** of that particular feature.
