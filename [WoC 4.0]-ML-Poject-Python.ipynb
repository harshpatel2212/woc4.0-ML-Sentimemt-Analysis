# LOOPS

**FOR**

for variable in something:

````python
for i in range(5):
    print(i)             #when starting point of range is not mentioned it is by default 0
````

````python
for i in range(5,20):
    print(i)             #when both starting and end of range is mentioned i runs for start to end-1   
````

````python
for i in range(5,20,3):
    print(i)             #if a third parameter besides start and end denotes the the increment in i after each iteration which is by default 1
````

**WHILE**

while some_condition:

````python
i = 1
while i < 3:
    print(i ** 2)
    i = i+1
print('MSTC WOC')          #while loops run until the condition does not get falsified
````

# LISTS

## Initializing a list

````python
l = []  #square brackets can be used to initialize a list 
print(type(l)) # output will show the datatype of variable 'l' as list
````

````python
l1=["hello","everyone"] # we can also initialize a list with some intial given values
print(l1)
````

````python
l2=["MSTC","WOC",4.0,True]  #single list can store more than 1 datatype 
print(l2)
````

## Add,update and access list elements

````python
l3 = [1,2,3,4]
l3.append(5)  # append operation add an element right after the last element in a list
print(l3)
l3.insert(1,"MSTC") #insert operation takes two arguments index and value and the value is inserted at the given index
print(l3)
````

````python
# list elements can be obtained by their indexes in python the index systems works as follows
l4 =   [1 , 2 , 3 , 4 , 5] 
#index  0 , 1 , 2 , 3 , 4
#index -5, -4, -3, -2, -1   

# we can access list elements by any of the given indexes shown above
print(l4[0],l4[-5])
print(l4[2],l4[-3])
````

````python
# to access range of elements in list we can do following thing
l4[0:2]  #this [a:b] method will start from index a and goes upto index b in rightward direction
         #And prints all elements from index a to b-1
````

````python
l4[0:-1] # this will print all the elements inside the list except the last one
````

````python
l4[-5:3] # other examples
````

````python
l4[-4:-1]
````

````python
l4[2:] # when end point is not provided it iterate all elements from start inded upto last including last
````

````python
l4[:-2] # when start index is not provided it iterate all elements from start to the given last index
````

````python
#updating list elements
l5 = [1,2,3,4,5,6]
l5[2] = "Winter"
l5[3]="of"     #we can directly overwrite/update list elements by using index and assign it to a new value
l5[4]="code"
print(l5)
````

##iterating lists

````python
#using for loops
temp = [1,2,3,4,5,6,"winter","of","code"]
for i in range(len(temp)): # function len returns the legth or number of elements inside the list
 print(temp[i])
````

````python
#using range based for loops

for ele in temp:   # here the variable ele directly access the elements values in the list "temp" on each ieteration
  print(ele)
````

# Dictionary

## Initializing and accessing a dictionary

A dictionary is an unordered and mutable Python container that stores mappings of unique keys to values.

````python
dic = {}  # we can initialize a dictonary by curly brackets 
print(type(dic))
````

````python
dic1 = {1.00:'hello','a':'everyone',3.00:'welcome to WOC'}
# we can initialize dictionary like above shown where each a:b denotes a as key and b as value

#ACCESSING

# we can access the dictionary elements by their corresponding keys

print(dic1[1.00])
print(dic1['a'])
print(dic1[3.00])
````

## updating,deleting and inserting in dictionaries


````python
dic1['b'] = 'new_ele'    # to insert type new key and element
dic1['a'] = 'updateed_everyone'  # to update change the value for given key by accessing it
dic1.pop(1.00)  # pop method removes key-value pair by taking parameter as key

print(dic1)
````

## operations on dictionaries

````python
# .keys() methods returns list object containing all keys in the dictionary 
print(dic1.keys())
# .values() methods returns list object containing all values in the dictionary 
print(dic1.values())
# .keys() methods returns objects in form of tuple containing all items(key-value pairs) in the dictionary 
print(dic1.items())
# .lens() methods returns number of items in the dictionary
print(len(dic1))
````

## iterating in dictionaries


````python
for key in dic1.keys(): # printing keys using for loop
  print(key)
````

````python
for value in dic1.values(): # printing values using for loop
  print(value)
````

````python
for key,value in dic1.items(): # printing key value pairs using for loop
  print(key,value)
````
